```python
# This is an input class. Do not edit.
class BinaryTree:
    def __init__(self, value, left=None, right=None):
        self.value = value
        self.left = left
        self.right = right


def heightBalancedBinaryTree(tree):
    # Write your code here.
    target=[0]
    helper(tree,target)
    if(target[0]<=1):
        return True
    return False

def helper(tree,target):
    if tree is None:
        return 0
    else:
        if(tree.left is None and tree.right is None):
            return 1
        lh=helper(tree.left,target)
        rh=helper(tree.right,target)
        if abs(lh-rh)>target[0]:
            target[0]=abs(lh-rh)
        return max(lh,rh)+1
    return 0
```
